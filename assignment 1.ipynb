```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("DS - Assignment Part 1 data set.xlsx")
```

```python
not_important_columns = ['Transaction date','latitude','longitude']
```

```python
df1 = df.drop(not_important_columns,axis='columns')
```

```python
df1.shape
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1 = df1[df1['House size (sqft)']/df1['Number of bedrooms'] >300]
```

```python
df1.shape
```

```python
X = df1.drop('House price of unit area',axis='columns')
X.head()
```

```python
y = df1['House price of unit area']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
find_best_model_using_gridsearchcv(X,y)
```

```python
def predict():
    House_Age = float(input("House_Age : "))
    Distance_from_NearestMetro_Inkm = float(input("Distance_from_NearestMetro in Km :  "))
    No_of_Convinence_Stores = float(input("No_of_Convinence_Stores : "))
    No_of_Bedrooms = float(input("No_of_Bedrooms : "))
    House_size_sqft = float(input("House_size_sqft :"))

    data = np.array([[House_Age, Distance_from_NearestMetro_Inkm, No_of_Convinence_Stores, No_of_Bedrooms, House_size_sqft]])
    output = lr_clf.predict(data)
    print("Price for the unit area of the house is : ", output)
```
